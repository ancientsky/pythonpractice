#### 1. 請讀取cdc_info.tsv 檔案中的「疾病別」欄位資料，並將資料整理至一名為disease_ary 的list之中 (提示: 使用變數、檔案IO、字串處理方法)

```python
with open("cdc_info.tsv",'r') as file:
    disease_ary=[]
    for line in file:
        ## print(line.strip().split('\t') 檢視各行
        disease_ary.append(line.strip().split('\t'))
disease_ary
```

```python
disease_ary=[]
with open("cdc_info.tsv",'r') as file:
    for line in file.readlines():
        disease_type=line.split()[0]
        if disease_type != '疾病別':
            disease_ary.append(line.split()[0])
```

```python
disease_ary=[line.split() [0] for line in open("cdc_info.tsv",'r')][1:]
print(disease_ary)
```

#### 2. 請使用Python 統計「疾病別」欄位中各疾病出現的數量 (提示: 使用字典或Collections)

```python
#Dictionary
dic={}
for disease in disease_ary:
    if disease not in dic:
        dic[disease]=1
    else:
        dic[disease]+=1
dic

##set
disease_set = set(disease_ary)
for disease in disease_set:
    print(disease, disease_ary.count(disease))
    
    
## collections
from collections import Counter
c=Counter(disease_ary)
c
```

```python
b=[]
for a in disease_ary:
    if disease_ary.index(a)>0:
        b.append(a[0])
from collections import Counter
c=dict(Counter(b))
for k in c:
    print(k,c[k])
```
