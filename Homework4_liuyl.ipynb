### get all version 1

```python
import requests
from bs4 import BeautifulSoup
import pandas
from datetime import datetime


newsary=[]

for repyear in range(2007,2018):  ##report dated back to 1996
    res=requests.get('http://www.who.int/csr/don/archive/year/{}/en/'.format(repyear))
    res.encoding='utf-8'
    soup=BeautifulSoup(res.text,'html.parser')
    soup.text
    for news in soup.select('.auto_archive li'):
        dt = news.select('a')[0].text
        date=datetime.strptime(dt,'%d %B %Y').date()
        url= news.select('a')[0]['href'].replace('/entity','http://www.who.int').replace('index.html','')
        disease = news.select('.link_info')[0].text.split('–')[0].strip()
        country = news.select('.link_info')[0].text.split('–')[-1].strip()
        newsary.append({'連結':url,'標題':disease,'日期':date,'國家':country})
#        newsary.append({'連結':url,'標題':disease})
        
df = pandas.DataFrame(newsary)
df[[ '連結','標題','日期','國家']]
df.to_excel('whodon1996-2017.xlsx')

#df[[ '連結','標題']]




```

### get all version 2

```python
import requests
from bs4 import BeautifulSoup
import pandas
from datetime import datetime
def getPageDetail(url):
    res=requests.get(url)
    res.encoding='utf-8'
    soup=BeautifulSoup(res.text,'html.parser')
    temp_array=[]
    for news in soup.select('.auto_archive li'):
        dt = news.select('a')[0].text
        date=datetime.strptime(dt,'%d %B %Y').date()
        url= news.select('a')[0]['href'].replace('/entity','http://www.who.int').replace('index.html','')
        disease = news.select('span')[0].text.split('–')[0].strip()
        country = news.select('span')[0].text.split('–')[-1].strip()
        temp_array.append({'Link':url,'Disease':disease,'Date':date,'Country':country})
        df=pandas.DataFrame(temp_array)
    return df

    
```

```python
newsary=[]

for repyear in range(2017,2018):
    url='http://www.who.int/csr/don/archive/year/{}/en/'.format(repyear)    
    newsary.append(getPageDetail(url))
        
df = pandas.concat(newsary)
df
#df.to_excel('whodon1996-2017.xlsx')

```
