## 4. 統計各病徵出現的數目? (提示：使用dictionary )

```python
disease_dic = {}
for name in disease:
    if name not in disease_dic:
        disease_dic[name] = 1
    else:
        disease_dic[name]  += 1
```

```python
for name in disease_dic:
    print(name,disease_dic.get(name))
```

## 5. 回答出在所有病例中，哪個病徵出現最多次? (提示：使用dictionary 或 list 的統計方法)

```python
max_value=max(disease_dic.values())
for k,v in disease_dic.items(): 
    if v==max_value:
        print(k)
```
